### Extraction de Features

### **Objectif** : Transformer le texte nettoyé en vecteurs numériques pour le machine learning

#### **Méthodes** :
##### 1. TF-IDF (Term Frequency - Inverse Document Frequency)
##### 2. Analyse et visualisation des features

### 1. Imports des bibliothèques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import save_npz, load_npz
import pickle
from pathlib import Path
import warnings

warnings.filterwarnings('ignore')

print(" Imports réussis!")
```

### 2. Chargement des Données Nettoyées

```python
# Charger le dataset nettoyé
df = pd.read_csv('../data/processed/resume_cleaned.csv')

print(f" Dataset chargé:")
print(f"   - Nombre de CV: {len(df)}")
print(f"   - Colonnes: {df.columns.tolist()}")

# Vérifier la colonne de texte nettoyé
print(f"\n Exemple de texte nettoyé:")
print(df['cleaned_text'].iloc[0][:300] + "...")
```

### 3. Configuration du TF-IDF Vectorizer

```python
# Créer le vectorizer TF-IDF
vectorizer = TfidfVectorizer(
    max_features=5000,      # Garder les 5000 mots les plus importants
    ngram_range=(1, 2),     # Unigrams et bigrams
    min_df=2,               # Mot doit apparaître dans au moins 2 documents
    max_df=0.95,            # Mot ne doit pas apparaître dans plus de 95% des documents
    sublinear_tf=True,      # Utiliser le scaling sublinéaire pour tf
    use_idf=True,
    smooth_idf=True
)

print(" TF-IDF Vectorizer configuré:")
print(f"   - Max features: 5000")
print(f"   - N-grams: (1, 2) - Unigrams et bigrams")
print(f"   - Min document frequency: 2")
print(f"   - Max document frequency: 0.95")
```

### 4. Transformation du Texte en Features

```python
# Extraire les textes
texts = df['cleaned_text'].values

print(f"\n Transformation de {len(texts)} CV en features...")

# Fit et transform
X = vectorizer.fit_transform(texts)

print(f"\n Transformation terminée!")
print(f"   - Shape de la matrice: {X.shape}")
print(f"   - Nombre de CV: {X.shape[0]}")
print(f"   - Nombre de features: {X.shape[1]}")
print(f"   - Type: {type(X)}")
print(f"   - Sparsité: {(1.0 - X.nnz / (X.shape[0] * X.shape[1])) * 100:.2f}%")
print(f"   - Mémoire: {X.data.nbytes / 1024**2:.2f} MB")
```

### 5. Analyse des Features

```python
# Obtenir les noms des features
feature_names = vectorizer.get_feature_names_out()

print(f"\n FEATURES EXTRAITES:")
print(f"   Total: {len(feature_names)}")
print(f"\n Exemples de features (50 premières):")
for i in range(0, 50, 10):
    print(f"   {i+1:3d}-{i+10:3d}: {', '.join(feature_names[i:i+10])}")
```

### 6. Analyse des Features les Plus Importantes

```python
# Calculer l'importance globale de chaque feature (somme des scores TF-IDF)
feature_importance = np.array(X.sum(axis=0)).flatten()

# Créer un DataFrame pour l'analyse
df_features = pd.DataFrame({
    'feature': feature_names,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

print("\n TOP 30 FEATURES LES PLUS IMPORTANTES:")
print("="*60)
print(df_features.head(30).to_string(index=False))
```

### 7. Visualisation des Top Features

```python
# Visualiser les top 20 features
fig, ax = plt.subplots(figsize=(12, 8))

top_20 = df_features.head(20)
ax.barh(range(len(top_20)), top_20['importance'].values, color='skyblue', edgecolor='black')
ax.set_yticks(range(len(top_20)))
ax.set_yticklabels(top_20['feature'].values)
ax.set_xlabel('Importance Globale (Somme TF-IDF)', fontweight='bold')
ax.set_title('Top 20 Features les Plus Importantes', fontsize=14, fontweight='bold')
ax.invert_yaxis()
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('../outputs/plots/top_features.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graphique sauvegardé dans outputs/plots/top_features.png")
```

### 8. Distribution des Scores TF-IDF

```python
# Analyser la distribution des scores
non_zero_scores = X.data  # Seulement les valeurs non-nulles

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Histogramme
axes[0].hist(non_zero_scores, bins=50, color='lightcoral', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Score TF-IDF', fontweight='bold')
axes[0].set_ylabel('Fréquence', fontweight='bold')
axes[0].set_title('Distribution des Scores TF-IDF', fontsize=14, fontweight='bold')
axes[0].axvline(non_zero_scores.mean(), color='red', linestyle='--',
               label=f'Moyenne: {non_zero_scores.mean():.4f}')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Boxplot
axes[1].boxplot(non_zero_scores, vert=True)
axes[1].set_ylabel('Score TF-IDF', fontweight='bold')
axes[1].set_title('Boxplot des Scores TF-IDF', fontsize=14, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../outputs/plots/tfidf_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n STATISTIQUES DES SCORES TF-IDF:")
print(f"   Moyenne: {non_zero_scores.mean():.4f}")
print(f"   Médiane: {np.median(non_zero_scores):.4f}")
print(f"   Min: {non_zero_scores.min():.4f}")
print(f"   Max: {non_zero_scores.max():.4f}")
print(f"   Écart-type: {non_zero_scores.std():.4f}")
```

### 9. Analyse par Catégorie

```python
# Analyser les features importantes par catégorie
category_col = 'Category'
categories = df[category_col].unique()

print(f"\n ANALYSE PAR CATÉGORIE")
print(f"   Nombre de catégories: {len(categories)}")

# Prendre les top 5 catégories
top_5_categories = df[category_col].value_counts().head(5).index

print(f"\n Top 5 catégories:")
for i, cat in enumerate(top_5_categories, 1):
    count = len(df[df[category_col] == cat])
    print(f"   {i}. {cat}: {count} CV")
```

### 10. Features Caractéristiques par Catégorie

```python
# Pour chaque top catégorie, trouver les features les plus importantes
print(f"\n FEATURES CARACTÉRISTIQUES PAR CATÉGORIE:")
print("="*80)

for category in top_5_categories[:3]:  # Top 3 pour ne pas surcharger
    # Indices des CV de cette catégorie
    cat_indices = df[df[category_col] == category].index
    
    # Scores moyens pour cette catégorie
    cat_scores = X[cat_indices].mean(axis=0).A1
    
    # Top 10 features pour cette catégorie
    top_indices = cat_scores.argsort()[-10:][::-1]
    top_features = [(feature_names[i], cat_scores[i]) for i in top_indices]
    
    print(f"\n {category}:")
    for i, (feature, score) in enumerate(top_features, 1):
        print(f"   {i:2d}. {feature:30s} : {score:.4f}")
```

### 11. Réduction de Dimensionnalité (Optionnel - LSA)

```python
# Réduire à 100 dimensions avec SVD (Latent Semantic Analysis)
print("\n Réduction de dimensionnalité avec SVD...")

n_components = 100
svd = TruncatedSVD(n_components=n_components, random_state=42)
X_reduced = svd.fit_transform(X)

explained_variance = svd.explained_variance_ratio_.sum() * 100

print(f"\n Réduction terminée:")
print(f"   Dimensions originales: {X.shape[1]}")
print(f"   Nouvelles dimensions: {X_reduced.shape[1]}")
print(f"   Variance expliquée: {explained_variance:.2f}%")

# Visualiser la variance expliquée
plt.figure(figsize=(12, 6))
plt.plot(range(1, n_components + 1), svd.explained_variance_ratio_.cumsum() * 100,
        marker='o', linewidth=2, markersize=4)
plt.xlabel('Nombre de Composantes', fontweight='bold')
plt.ylabel('Variance Expliquée Cumulée (%)', fontweight='bold')
plt.title('Variance Expliquée par les Composantes SVD', fontsize=14, fontweight='bold')
plt.grid(alpha=0.3)
plt.axhline(y=80, color='r', linestyle='--', label='80%')
plt.axhline(y=90, color='g', linestyle='--', label='90%')
plt.legend()
plt.tight_layout()
plt.savefig('../outputs/plots/svd_variance.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graphique sauvegardé")
```

### 12. Sauvegarde des Features et du Vectorizer

```python
# Créer le dossier si nécessaire
Path('../data/processed').mkdir(parents=True, exist_ok=True)
Path('../models_saved').mkdir(parents=True, exist_ok=True)

# Sauvegarder la matrice TF-IDF (format sparse)
save_npz('../data/processed/X_features.npz', X)
print(" Matrice TF-IDF sauvegardée: data/processed/X_features.npz")

# Sauvegarder le vectorizer
with open('../models_saved/tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorizer, f)
print(" Vectorizer sauvegardé: models_saved/tfidf_vectorizer.pkl")

# Sauvegarder aussi la version réduite (optionnel)
np.save('../data/processed/X_features_reduced.npy', X_reduced)
print(" Features réduites (SVD) sauvegardées: data/processed/X_features_reduced.npy")

# Sauvegarder le modèle SVD
with open('../models_saved/svd_model.pkl', 'wb') as f:
    pickle.dump(svd, f)
print(" Modèle SVD sauvegardé: models_saved/svd_model.pkl")
```

### 13. Test de Rechargement

```python
# Tester le rechargement
print("\n TEST DE RECHARGEMENT:")
print("="*60)

# Recharger la matrice
X_loaded = load_npz('../data/processed/X_features.npz')
print(f" Matrice rechargée: {X_loaded.shape}")

# Recharger le vectorizer
with open('../models_saved/tfidf_vectorizer.pkl', 'rb') as f:
    vectorizer_loaded = pickle.load(f)
print(f" Vectorizer rechargé: {vectorizer_loaded.max_features} features")

# Test de transformation sur un nouveau texte
test_text = "python developer machine learning data science artificial intelligence"
test_vector = vectorizer_loaded.transform([test_text])
print(f" Test de transformation: {test_vector.shape}")
print(f"   Valeurs non-nulles: {test_vector.nnz}")
```
