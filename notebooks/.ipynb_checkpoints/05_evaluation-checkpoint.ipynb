### 1. Imports et Chargement

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import load_npz
import pickle
from sklearn.metrics import (classification_report, confusion_matrix,
                            accuracy_score, precision_recall_fscore_support)
import warnings

warnings.filterwarnings('ignore')

print(" Imports réussis!")
```

```python
# Charger les données
df = pd.read_csv('../data/processed/resume_cleaned.csv')
X = load_npz('../data/processed/X_features.npz')

# Charger le meilleur modèle
with open('../models_saved/best_model.pkl', 'rb') as f:
    best_model = pickle.load(f)

# Charger le label encoder
with open('../models_saved/label_encoder.pkl', 'rb') as f:
    label_encoder = pickle.load(f)

# Charger le vectorizer
with open('../models_saved/tfidf_vectorizer.pkl', 'rb') as f:
    vectorizer = pickle.load(f)

print(f" Modèles et données chargés:")
print(f"   - CV: {len(df)}")
print(f"   - Features: {X.shape[1]}")
print(f"   - Modèle: {type(best_model).__name__}")
print(f"   - Catégories: {len(label_encoder.classes_)}")
```

### 2. Split Train/Test

```python
from sklearn.model_selection import train_test_split

# Préparer les données
category_col = 'Category'
y = df[category_col].values
y_encoded = label_encoder.transform(y)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f" Split des données:")
print(f"   Train: {len(y_train)} ({len(y_train)/len(y)*100:.1f}%)")
print(f"   Test:  {len(y_test)} ({len(y_test)/len(y)*100:.1f}%)")
```

### 3. Prédictions

```python
# Prédictions
y_pred_train = best_model.predict(X_train)
y_pred_test = best_model.predict(X_test)

# Prédictions probabilistes
y_prob_test = best_model.predict_proba(X_test)

print(" Prédictions effectuées")
```

### 4. Métriques Globales

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculer les métriques
train_acc = accuracy_score(y_train, y_pred_train)
test_acc = accuracy_score(y_test, y_pred_test)
precision = precision_score(y_test, y_pred_test, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred_test, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred_test, average='weighted', zero_division=0)

print("\n" + "="*80)
print(" MÉTRIQUES GLOBALES")
print("="*80)
print(f"""
   Train Accuracy:  {train_acc:.4f}
   Test Accuracy:   {test_acc:.4f}
   Precision:       {precision:.4f}
   Recall:          {recall:.4f}
   F1-Score:        {f1:.4f}
   
   Overfitting: {(train_acc - test_acc):.4f} ({'Oui' if train_acc - test_acc > 0.1 else 'Non'})
""")
```

### 5. Rapport de Classification Détaillé

```python
# Rapport complet
print("\n" + "="*80)
print(" RAPPORT DE CLASSIFICATION DÉTAILLÉ")
print("="*80)

report = classification_report(
    y_test, y_pred_test,
    target_names=label_encoder.classes_,
    digits=3
)
print(report)

# Sauvegarder
with open('../outputs/reports/final_evaluation_report.txt', 'w') as f:
    f.write("RAPPORT D'ÉVALUATION FINALE\n")
    f.write("="*80 + "\n\n")
    f.write(f"Modèle: {type(best_model).__name__}\n")
    f.write(f"Test Accuracy: {test_acc:.4f}\n")
    f.write(f"F1-Score: {f1:.4f}\n\n")
    f.write(report)

print("\n Rapport sauvegardé dans outputs/reports/final_evaluation_report.txt")
```

### 6. Matrice de Confusion

```python
# Calculer la confusion matrix
cm = confusion_matrix(y_test, y_pred_test)

# Visualiser (top 15 catégories)
n_classes = min(15, len(label_encoder.classes_))
cm_display = cm[:n_classes, :n_classes]
classes_display = label_encoder.classes_[:n_classes]

plt.figure(figsize=(16, 14))
sns.heatmap(cm_display, annot=True, fmt='d', cmap='Blues',
           xticklabels=classes_display,
           yticklabels=classes_display,
           cbar_kws={'label': 'Nombre de prédictions'})

plt.title(f'Confusion Matrix - {type(best_model).__name__}\n(Top {n_classes} catégories)', 
         fontsize=16, fontweight='bold')
plt.ylabel('True Label', fontweight='bold', fontsize=12)
plt.xlabel('Predicted Label', fontweight='bold', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('../outputs/plots/final_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Confusion matrix sauvegardée")
```

### 7. Analyse des Erreurs

```python
# Identifier les erreurs
errors_mask = y_test != y_pred_test
error_indices = np.where(errors_mask)[0]

print(f"\n ANALYSE DES ERREURS:")
print(f"   Total d'erreurs: {len(error_indices)}")
print(f"   Taux d'erreur: {len(error_indices)/len(y_test)*100:.2f}%")

# Analyser les paires d'erreurs les plus fréquentes
error_pairs = list(zip(y_test[errors_mask], y_pred_test[errors_mask]))
from collections import Counter
error_counter = Counter(error_pairs)

print(f"\n Top 10 Erreurs les Plus Fréquentes:")
print("="*80)
print(f"{'Vraie Catégorie':<30} {'Prédite Comme':<30} {'Occurrences':>12}")
print("-"*80)
for (true_idx, pred_idx), count in error_counter.most_common(10):
    true_cat = label_encoder.inverse_transform([true_idx])[0]
    pred_cat = label_encoder.inverse_transform([pred_idx])[0]
    print(f"{true_cat:<30} {pred_cat:<30} {count:>12}")
```

### 8. Performance par Catégorie

```python
# Calculer les métriques par classe
from sklearn.metrics import precision_recall_fscore_support

precisions, recalls, f1s, supports = precision_recall_fscore_support(
    y_test, y_pred_test, zero_division=0
)

# Créer un DataFrame
df_perf = pd.DataFrame({
    'Category': label_encoder.classes_,
    'Precision': precisions,
    'Recall': recalls,
    'F1-Score': f1s,
    'Support': supports
}).sort_values('F1-Score', ascending=False)

print("\n PERFORMANCE PAR CATÉGORIE (Top 15):")
print("="*80)
print(df_perf.head(15).to_string(index=False))

# Visualiser
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# Top 10 meilleures catégories
top_10 = df_perf.head(10)
x = np.arange(len(top_10))
width = 0.25

bars1 = axes[0].bar(x - width, top_10['Precision'], width, label='Precision', color='skyblue')
bars2 = axes[0].bar(x, top_10['Recall'], width, label='Recall', color='lightcoral')
bars3 = axes[0].bar(x + width, top_10['F1-Score'], width, label='F1-Score', color='lightgreen')

axes[0].set_xlabel('Catégorie', fontweight='bold')
axes[0].set_ylabel('Score', fontweight='bold')
axes[0].set_title('Top 10 Catégories - Meilleure Performance', fontsize=14, fontweight='bold')
axes[0].set_xticks(x)
axes[0].set_xticklabels(top_10['Category'], rotation=45, ha='right')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)
axes[0].set_ylim([0, 1.1])

# Bottom 10 catégories
bottom_10 = df_perf.tail(10).sort_values('F1-Score')
x = np.arange(len(bottom_10))

bars1 = axes[1].bar(x - width, bottom_10['Precision'], width, label='Precision', color='skyblue')
bars2 = axes[1].bar(x, bottom_10['Recall'], width, label='Recall', color='lightcoral')
bars3 = axes[1].bar(x + width, bottom_10['F1-Score'], width, label='F1-Score', color='lightgreen')

axes[1].set_xlabel('Catégorie', fontweight='bold')
axes[1].set_ylabel('Score', fontweight='bold')
axes[1].set_title('Bottom 10 Catégories - Moins Bonne Performance', fontsize=14, fontweight='bold')
axes[1].set_xticks(x)
axes[1].set_xticklabels(bottom_10['Category'], rotation=45, ha='right')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)
axes[1].set_ylim([0, 1.1])

plt.tight_layout()
plt.savefig('../outputs/plots/performance_by_category.png', dpi=300, bbox_inches='tight')
plt.show()

print(" Graphiques sauvegardés")
```

### 9. Analyse de Confiance des Prédictions

```python
# Analyser la confiance
confidences = y_prob_test.max(axis=1)

print(f"\n ANALYSE DE LA CONFIANCE:")
print(f"   Confiance moyenne: {confidences.mean():.4f}")
print(f"   Confiance médiane: {np.median(confidences):.4f}")
print(f"   Confiance min: {confidences.min():.4f}")
print(f"   Confiance max: {confidences.max():.4f}")

# Visualiser
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Distribution globale
axes[0].hist(confidences, bins=50, color='purple', edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Confiance', fontweight='bold')
axes[0].set_ylabel('Fréquence', fontweight='bold')
axes[0].set_title('Distribution de la Confiance des Prédictions', fontsize=14, fontweight='bold')
axes[0].axvline(confidences.mean(), color='red', linestyle='--',
               label=f'Moyenne: {confidences.mean():.3f}')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Confiance pour prédictions correctes vs incorrectes
correct_conf = confidences[~errors_mask]
incorrect_conf = confidences[errors_mask]

axes[1].boxplot([correct_conf, incorrect_conf],
               labels=['Correctes', 'Incorrectes'])
axes[1].set_ylabel('Confiance', fontweight='bold')
axes[1].set_title('Confiance: Prédictions Correctes vs Incorrectes', fontsize=14, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../outputs/plots/confidence_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n   Confiance moyenne (correctes): {correct_conf.mean():.4f}")
print(f"   Confiance moyenne (incorrectes): {incorrect_conf.mean():.4f}")
```

### 10. Exemples de Prédictions

```python
# Afficher quelques exemples
print("\n" + "="*80)
print(" EXEMPLES DE PRÉDICTIONS")
print("="*80)

# Quelques indices aléatoires du test set
sample_indices = np.random.choice(len(y_test), size=5, replace=False)

for i, idx in enumerate(sample_indices, 1):
    true_label = label_encoder.inverse_transform([y_test[idx]])[0]
    pred_label = label_encoder.inverse_transform([y_pred_test[idx]])[0]
    confidence = y_prob_test[idx].max()
    is_correct = "Correcte" if true_label == pred_label else "Incorrecte"
    
    # Obtenir le texte (correspondre idx du test set à idx du df original)
    original_idx = X_test[idx].indices[0] if hasattr(X_test[idx], 'indices') else idx
    
    print(f"\n{is_correct} Exemple {i}:")
    print(f"   Vraie catégorie:     {true_label}")
    print(f"   Catégorie prédite:   {pred_label}")
    print(f"   Confiance:           {confidence:.2%}")
```

```python
print("\n" + "="*80)
print(" RÉSUMÉ DE L'ÉVALUATION FINALE")
print("="*80)

print(f"""
 MODÈLE: {type(best_model).__name__}

 PERFORMANCE GLOBALE:
   Test Accuracy:  {test_acc:.4f}
   Precision:      {precision:.4f}
   Recall:         {recall:.4f}
   F1-Score:       {f1:.4f}

 ERREURS:
   Total: {len(error_indices)} sur {len(y_test)}
   Taux d'erreur: {len(error_indices)/len(y_test)*100:.2f}%

 CONFIANCE:
   Moyenne: {confidences.mean():.4f}
   Correctes: {correct_conf.mean():.4f}
   Incorrectes: {incorrect_conf.mean():.4f}

 MEILLEURES CATÉGORIES (F1-Score):
   1. {df_perf.iloc[0]['Category']}: {df_perf.iloc[0]['F1-Score']:.4f}
   2. {df_perf.iloc[1]['Category']}: {df_perf.iloc[1]['F1-Score']:.4f}
   3. {df_perf.iloc[2]['Category']}: {df_perf.iloc[2]['F1-Score']:.4f}

 FICHIERS GÉNÉRÉS:
   - outputs/reports/final_evaluation_report.txt
   - outputs/plots/final_confusion_matrix.png
   - outputs/plots/performance_by_category.png
   - outputs/plots/confidence_analysis.png

 LE MODÈLE EST PRÊT POUR LA PRODUCTION!
""")
```
